```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv("data.csv")
df.head()
```

```python
df.describe(include='all')
```

```python
df.drop(columns=['Unnamed: 0', 'linearAcX', 'linearAcY', 'linearAcZ', 'orientX', 'orientY'],inplace=True)
df.head(15)
```

```python
drivers= df['class'].unique()
```

```python
df.set_index(['class'],inplace=True)
df.head()
```

```python
for i in  drivers:
    temp=df.loc[i]
    filename=f'{i}_data.csv'
    temp.to_csv(filename)
```

```python
berk= pd.read_csv('berk_data.csv')
apo= pd.read_csv('apo_data.csv')
hurcan= pd.read_csv('hurcan_data.csv')
mehdi= pd.read_csv('mehdi_data.csv')
onder= pd.read_csv('onder_data.csv')
selin= pd.read_csv('selin_data.csv')
```

```python
plt.figure(figsize=(16, 12))
plt.plot(berk['accelZ'][0:20000], color= 'black', label='Acceleration along Z axis of Berk')
plt.plot(apo['accelZ'][0:20000], color= 'red', label='Acceleration along Z axis of apo')
plt.plot(hurcan['accelZ'][0:20000], color= 'green', label='Acceleration along Z axis of hurcan')
plt.plot(mehdi['accelZ'][0:20000], color= 'blue', label='Acceleration along Z axis of mehdi')
plt.plot(onder['accelZ'][0:20000], color= 'violet', label='Acceleration along Z axis of onder')
plt.plot(selin['accelZ'][0:20000], color= 'orange', label='Acceleration along Z axis of selin')
plt.title("Accelration along Z of different drivers")
plt.xlabel("Time")
plt.ylabel("Accelration along Z")
plt.legend()
plt.show()
```
